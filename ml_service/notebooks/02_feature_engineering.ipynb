###  Pressure Features

```python
def engineer_pressure_features(df, scenario_col="label", dt=0.5, epsilon=1e-8):

    df = df.sort_values([scenario_col, "timestep"]).copy()

    # spatial gradients
    df["pressure_drop_ab"] = df["node_a_pressure"] - df["node_b_pressure"]
    df["pressure_drop_bc"] = df["node_b_pressure"] - df["node_c_pressure"]
    df["pressure_drop_ac"] = df["node_a_pressure"] - df["node_c_pressure"]

    df["pressure_gradient_ab"] = df["pressure_drop_ab"] / 20.0
    df["pressure_gradient_bc"] = df["pressure_drop_bc"] / 20.0
    df["pressure_gradient_ac"] = df["pressure_drop_ac"] / 40.0

    # midpoint analysis
    """
    positive = blockage (pressure spike at B)
    negative = leak (pressure drop at B)
    """
    df["expected_midpoint_pressure"] = (df["node_a_pressure"] + df["node_c_pressure"])/2
    df["midpoint_deviation"] = df["node_b_pressure"] - df["expected_midpoint_pressure"]

    # Pressure asymmetry
    df["pressure_asynnetry"] = abs(df["pressure_drop_ab"] -  df["pressure_drop_bc"])

    # Normalized pressure tactics
    df["pressure_ratio_ab"] = df["node_a_pressure"] / (df["node_b_pressure"] + epsilon)
    df["pressure_ratio_bc"] = df["node_b_pressure"] / (df["node_c_pressure"] + epsilon)
    df["pressure_ratio_ac"] = df["node_a_pressure"] / (df["node_c_pressure"] + epsilon)

    # temporal rates (per scenario)



    return df
```

###  Velocity Features

```python
def engineer_velocity_features(df):

    df['velocity_drop_ab'] = df['velocity_a'] - df['velocity_b']
    df['velocity_drop_bc'] = df['velocity_b'] - df['velocity_c']
    df['velocity_drop_ac'] = df['velocity_a'] - df['velocity_c']

    df['mean_velocity'] = df[['velocity_a', 'velocity_b', 'velocity_c']].mean(axis=1)

    df['velocity_deviation_a'] = df['velocity_a'] - df['mean_velocity']
    df['velocity_deviation_b'] = df['velocity_b'] - df['mean_velocity']
    df['velocity_deviation_c'] = df['velocity_c'] - df['mean_velocity']

    # for each scenario group sorted by timestep:
    df = df.sort_values(['label', 'timestep']).copy()

    df['dv_dt_a'] = df.groupby('label')['velocity_a'].diff() / 0.5
    df['dv_dt_b'] = df.groupby('label')['velocity_b'].diff() / 0.5
    df['dv_dt_c'] = df.groupby('label')['velocity_c'].diff() / 0.5

    df[['dv_dt_a', 'dv_dt_b', 'dv_dt_c']] = df[['dv_dt_a', 'dv_dt_b', 'dv_dt_c']].fillna(0)

    return df
```
