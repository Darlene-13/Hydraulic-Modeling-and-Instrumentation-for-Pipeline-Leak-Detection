###  Pressure Features

```python
def engineer_pressure_features(df, scenario_col="label", dt=0.5, epsilon=1e-8):

    df = df.sort_values([scenario_col, "timestep"]).copy()

    # spatial gradients
    df["pressure_drop_ab"] = df["node_a_pressure"] - df["node_b_pressure"]
    df["pressure_drop_bc"] = df["node_b_pressure"] - df["node_c_pressure"]
    df["pressure_drop_ac"] = df["node_a_pressure"] - df["node_c_pressure"]

    distance_ab = 20.0
    distance_bc = 20.0
    distance_ac = 40.0

    df["pressure_gradient_ab"] = df["pressure_drop_ab"] / distance_ab
    df["pressure_gradient_bc"] = df["pressure_drop_bc"] / distance_bc
    df["pressure_gradient_ac"] = df["pressure_drop_ac"] / distance_ac

    # midpoint analysis
    """
    positive = blockage (pressure spike at B)
    negative = leak (pressure drop at B)
    """
    df["expected_midpoint_pressure"] = (df["node_a_pressure"] + df["node_c_pressure"])/2
    df["midpoint_pressure_deviation"] = df["node_b_pressure"] - df["expected_midpoint_pressure"]

    # Pressure asymmetry
    # Measures how lopsided the pressure profile is: expected large asymmetry to blockages because AB would drop low and BC drop high, while leak would be moderate asymmetry.
    df["pressure_asymmetry"] = abs(df["pressure_drop_ab"] -  df["pressure_drop_bc"])

    # Normalized pressure tactics
    df["pressure_ratio_ab"] = df["node_a_pressure"] / (df["node_b_pressure"] + epsilon)
    df["pressure_ratio_bc"] = df["node_b_pressure"] / (df["node_c_pressure"] + epsilon)
    df["pressure_ratio_ac"] = df["node_a_pressure"] / (df["node_c_pressure"] + epsilon)

    # temporal rates (per scenario)
    def compute_temporal_groups(group):
        dt = 0.5  # seconds between timesteps

        group["dp_dt_a"] = group["node_a_pressure"].diff() / dt
        group["dp_dt_b"] = group["node_b_pressure"].diff() / dt
        group["dp_dt_c"] = group["node_c_pressure"].diff() / dt

        # Second derivatives
        group["d2p_dt2_a"] = group["dp_dt_a"].diff() / dt
        group["d2p_dt2_b"] = group["dp_dt_b"].diff() / dt

        return group

    df = df.groupby("scenario_id").apply(compute_temporal_groups)

    # fill first timestep with zero

    temporal_cols = ["dp_dt_a", "dp_dt_b", "dp_dt_c", "d2p_dt2_a", "d2p_dt2_b"]
    df[temporal_cols] = df[temporal_cols].fillna(0.0)

    return df


# Calling to check the shape of the new dataframe with pressure features
df = engineer_pressure_features(df)
print(f"New shape after pressure feature engineering: {df.shape}")
```

###  Velocity Features

```python
def engineer_velocity_features(df):
    # Spatial differences
    df['velocity_drop_ab'] = df['velocity_a'] - df['velocity_b']
    df['velocity_drop_bc'] = df['velocity_b'] - df['velocity_c']
    df['velocity_drop_ac'] = df['velocity_a'] - df['velocity_c']

    # Mean and deviation
    df['mean_velocity'] = df[['velocity_a', 'velocity_b', 'velocity_c']].mean(axis=1)

    df["velocity_deviation_a"] = df["velocity_a"] - df["mean_velocity"]
    df["velocity_deviation_b"] = df["velocity_b"] - df["mean_velocity"]
    df["velocity_deviation_c"] = df["velocity_c"] - df["mean_velocity"]

    df["velocity_std"] = df[["velocity_a", "velocity_b", "velocity_c"]].std(axis=1)

    # Midpoint velocity deviation
    df["expected_midpoint_velocity"] = (df["velocity_a"] + df["velocity_c"]) / 2
    df["midpoint_velocity_deviation"] = df["velocity_b"] - df["expected_midpoint_velocity"]

    # Temporal rates of change
    def compute_velocity_temporal_groups(group):
        dt = 0.5 # Second between timesteps

        group["dv_dt_a"] = group['velocity_a'].diff() / dt
        group["dv_dt_b"] = group['velocity_b'].diff() / dt
        group["dv_dt_c"] = group['velocity_c'].diff() / dt

        # Second derivatives
        group["d2v_dt2_a"] = group["dv_dt_a"].diff() / dt
        group["d2v_dt2_b"] = group["dv_dt_b"].diff() /dt
        group["d2v_dt2_c"] = group["dv_dt_c"].diff() / dt

        return group

    # for each scenario group sorted by timestep:
    df = df.groupby("scenario_id").apply(compute_velocity_temporal_groups)

    # fill first timestep with zero
    temporal_cols = ["dv_dt_a", "dv_dt_b", "dv_dt_c", "d2v_dt2_a", "d2v_dt2_b", "d2v_dt2_c"]
    df[temporal_cols] = df[temporal_cols].fillna(0.0)

    return df

df = engineer_velocity_features(df)
print(f"New shape after pressure feature engineering: {df.shape}")
```

### Turbulence Features

```python
# Turbulence Kinetic velocity features
def engineer_turbulence_features(df, epsilon = 1e-8):

    # Turbulence intensity per node
    df["turb_intensity_a"] = np.sqrt((2.0 / 3.0) * df["tke_a"]) / (df["velocity_a"] + epsilon)
    df["turb_intensiy_b"] = np.sqrt((2.0 / 3.0) * df["tke_b"]) / (df["velocity_b"] + epsilon)
    df["turb_intensity_c"] = np.sqrt((2.0 / 3.0) * df["tke_c"]) / (df["velocity_c"] + epsilon)

    # Spatial tke differences
    df["tke_drop_ab"] = df["tke_a"] - df["tke_b"]
    df["tke_drop_bc"] = df["tke_b"] - df["tke_c"]
    df["tke_drop_ac"] = df["tke_a"] - df["tke_c"]

    # Dissipation ratios
    df["tke_tdr_ratio_a"] = df["tke_a"] / (df["tdr_a"] + epsilon)
    df["tke_tdr_ratio_b"] = df["tke_b"] / (df["tdr_b"] + epsilon)
    df["tke_tdr_ratio_c"] = df["tke_c"] / (df["tdr_c"] + epsilon)

    # Turbulence length scale proxy
    df["turb_length_a"] = (df["tke_a"] ** 1.5) / (df["tdr_a"] + epsilon)
    df["turb_length_b"] = (df["tke_b"] ** 1.5) / (df["tdr_b"] + epsilon)
    df["turb_length_c"] = (df["tke_c"] ** 1.5) / (df["tdr_c"] + epsilon)

    return df

df = engineer_turbulence_features(df)
print(f"New shape after turbulence feature engineering: {df.shape}")
```

### Wall Shear Engineering

```python
def engineer_wall_shear_features(df,epsilon = 1e-8):

    # Spatial differences
    df["shear_drop_ab"] = df["wall_shear_a"] - df["wall_shear_b"]
    df["shear_drop_ac"] = df["wall_shear_a"] - df["wall_shear_c"]
    df["shear_drop_bc"] = df["wall_shear_b"] - df["wall_shear_c"]

    # Shear to velocity ratio
    df["shear_vel_ratio_a"] = df["wall_shear_a"] / (df["velocity_a"] + epsilon)
    df["shear_vel_ratio_b"] = df["wall_shear_b"] / (df["velocity_b"] + epsilon)
    df["shear_vel_ratio_c"] = df["wall_shear_c"] / (df["velocity_c"] + epsilon)

    # total and mean shear
    df["total_shear"] = df["wall_shear_a"] + df["wall_shear_b"] + df["wall_shear_c"]
    df["mean_shear"] = df["total_shear"] / 3
    df["shear_std"] = df[["wall_shear_a", "wall_shear_b", "wall_shear_c"]].std(axis=1)

    # Shear assymmetry
    df["shear_asymmetry"] = abs(df["shear_drop_ab"] - df["shear_drop_bc"])

    return df

df = engineer_wall_shear_features(df)
print(f"New shape after wall shear feature engineering: {df.shape}")
```

### Tailings VOF Engineering

```python
def engineer_vof_features(df):

    distance_ac = 40.0  # distance between node A and C in meters

    # Spatial differences
    df["vof_drop_ab"] = df["tailings_vof_a"] - df["tailings_vof_b"]
    df["vof_drop_ac"] = df["tailings_vof_a"] - df["tailings_vof_c"]
    df["vof_drop_bc"] = df["tailings_vof_b"] - df["tailings_vof_c"]

    # Mean and Deviation
    df["mean_vof"] = df[["tailings_vof_a", "tailings_vof_b", "tailings_vof_c"]].mean(axis=1)
    df["vof_std"] = df[["tailings_vof_a", "tailings_vof_b", "tailings_vof_c"]].std(axis=1)

    # Concentration gradient along the pipe
    df["vof_gradient"] = (df["tailings_vof_c"] - df["tailings_vof_a"]) / distance_ac

    # Solid flux per node
    df["solid_flux_a"] = df["tailings_vof_a"] * df["velocity_a"]
    df["solid_flux_b"] = df["tailings_vof_b"] * df["velocity_b"]
    df["solid_flux_c"] = df["tailings_vof_c"] * df["velocity_c"]

    # Solid flux differences
    df["flux_drop_ab"] = df["solid_flux_a"] - df["solid_flux_b"]
    df["flux_drop_bc"] = df["solid_flux_b"] - df["solid_flux_c"]
    df["flux_drop_ac"] = df["solid_flux_a"] - df["solid_flux_c"]

    # Midpoint of vof  deviation
    df["expected_midpoint_vof"] = (df["tailings_vof_a"] + df["tailings_vof_c"]) / 2
    df["midpoint_vof_deviation"] = df["tailings_vof_b"] - df["expected_midpoint_vof"]

    return df

df = engineer_vof_features(df)
print(f"New shape after tailings VOF feature engineering: {df.shape}")
```
